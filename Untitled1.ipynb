```python
import pandas as pd
import numpy as np

# ==========================================
# ⚙️ الإعدادات
# ==========================================
INPUT_FILE  = "insurance.csv"
OUTPUT_FILE = "insurance_cleaned.csv"

# ==========================================
# 📥 قراءة البيانات
# ==========================================
df = pd.read_csv(INPUT_FILE)

print("="*50)
print("📊 البيانات قبل التنظيف")
print("="*50)
print(f"الأبعاد: {df.shape[0]} صف × {df.shape[1]} عمود")
print("\nأول 5 صفوف:")
print(df.head())
print("\nمعلومات الأعمدة:")
print(df.info())
print("\nقيم ناقصة:")
print(df.isna().sum())


# ==========================================
# 🔤 الخطوة 1: تنظيف أسماء الأعمدة
# ==========================================
# إزالة أي مسافات أو رموز غريبة في أسماء الأعمدة
df.columns = df.columns.str.strip().str.lower().str.replace(r"[^\w]", "", regex=True)

print("\n✅ أسماء الأعمدة بعد التنظيف:", list(df.columns))


# ==========================================
# 🔁 الخطوة 2: حذف الصفوف المكررة
# ==========================================
before = len(df)
df.drop_duplicates(inplace=True)
print(f"\n✅ تم حذف {before - len(df)} صف مكرر")


# ==========================================
# 🔢 الخطوة 3: تنظيف عمود age (السن)
# ==========================================
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# السن لازم يكون بين 0 و 120
invalid_age = df[(df['age'] < 0) | (df['age'] > 120)].shape[0]
df.loc[(df['age'] < 0) | (df['age'] > 120), 'age'] = np.nan

# ملء القيم الناقصة بالوسيط
median_age = df['age'].median()
df['age'].fillna(median_age, inplace=True)
df['age'] = df['age'].astype(int)

print(f"\n✅ عمود age: تم تصحيح {invalid_age} قيمة شاذة، الوسيط = {median_age:.0f}")


# ==========================================
# 👤 الخطوة 4: تنظيف عمود sex (الجنس)
# ==========================================
df['sex'] = df['sex'].str.strip().str.lower()

# توحيد القيم المختلفة
sex_mapping = {
    'm': 'male', 'man': 'male', 'ذكر': 'male',
    'f': 'female', 'woman': 'female', 'أنثى': 'female'
}
df['sex'] = df['sex'].replace(sex_mapping)

# أي قيمة مش male أو female → unknown
df.loc[~df['sex'].isin(['male', 'female']), 'sex'] = 'unknown'
df['sex'].fillna('unknown', inplace=True)

print(f"\n✅ عمود sex: {df['sex'].value_counts().to_dict()}")


# ==========================================
# 📏 الخطوة 5: تنظيف عمود bmi
# ==========================================
df['bmi'] = pd.to_numeric(df['bmi'], errors='coerce')

# BMI الطبيعي بين 10 و 60
invalid_bmi = df[(df['bmi'] < 10) | (df['bmi'] > 60)].shape[0]
df.loc[(df['bmi'] < 10) | (df['bmi'] > 60), 'bmi'] = np.nan

median_bmi = df['bmi'].median()
df['bmi'].fillna(median_bmi, inplace=True)
df['bmi'] = df['bmi'].round(2)

print(f"\n✅ عمود bmi: تم تصحيح {invalid_bmi} قيمة شاذة، الوسيط = {median_bmi:.2f}")


# ==========================================
# 👶 الخطوة 6: تنظيف عمود children (الأولاد)
# ==========================================
df['children'] = pd.to_numeric(df['children'], errors='coerce')

# عدد الأولاد لازم يكون بين 0 و 10
df.loc[(df['children'] < 0) | (df['children'] > 10), 'children'] = np.nan
df['children'].fillna(0, inplace=True)
df['children'] = df['children'].astype(int)

print(f"\n✅ عمود children: {df['children'].value_counts().sort_index().to_dict()}")


# ==========================================
# 🚬 الخطوة 7: تنظيف عمود smoker (مدخن)
# ==========================================
df['smoker'] = df['smoker'].str.strip().str.lower()

# توحيد القيم
smoker_mapping = {
    'yes': 'yes', 'y': 'yes', '1': 'yes', 'true': 'yes', 'مدخن': 'yes',
    'no': 'no',  'n': 'no',  '0': 'no',  'false': 'no', 'غير مدخن': 'no'
}
df['smoker'] = df['smoker'].replace(smoker_mapping)
df.loc[~df['smoker'].isin(['yes', 'no']), 'smoker'] = 'unknown'
df['smoker'].fillna('unknown', inplace=True)

print(f"\n✅ عمود smoker: {df['smoker'].value_counts().to_dict()}")


# ==========================================
# 🌍 الخطوة 8: تنظيف عمود region (المنطقة)
# ==========================================
df['region'] = df['region'].str.strip().str.lower().str.replace(r"[^\w\s]", "", regex=True)
df['region'].fillna('unknown', inplace=True)

print(f"\n✅ عمود region: {df['region'].value_counts().to_dict()}")


# ==========================================
# 💰 الخطوة 9: تنظيف عمود expenses (المصاريف)
# ==========================================
df['expenses'] = pd.to_numeric(df['expenses'], errors='coerce')

# المصاريف لازم تكون أكبر من صفر
df.loc[df['expenses'] <= 0, 'expenses'] = np.nan

median_exp = df['expenses'].median()
df['expenses'].fillna(median_exp, inplace=True)
df['expenses'] = df['expenses'].round(2)

print(f"\n✅ عمود expenses: الوسيط = {median_exp:.2f}")


# ==========================================
# 📊 تقرير ما بعد التنظيف
# ==========================================
print("\n" + "="*50)
print("📊 البيانات بعد التنظيف")
print("="*50)
print(f"الأبعاد: {df.shape[0]} صف × {df.shape[1]} عمود")
print(f"قيم ناقصة متبقية: {df.isna().sum().sum()}")
print("\nإحصائيات:")
print(df.describe())


# ==========================================
# 💾 حفظ البيانات النظيفة
# ==========================================
df.to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
print(f"\n💾 تم الحفظ في: {OUTPUT_FILE}")
print("\n🎉 تم التنظيف بنجاح!")
```
